필기체숫자사진받아서 각 숫자 라벨링

```python
import cv2
import numpy as np

src = cv2.imread('number.jpeg', cv2.IMREAD_GRAYSCALE)
_, src_bin = cv2.threshold(src, 0, 255, cv2.THRESH_OTSU)
cnt, labels, stats, centroids = cv2.connectedComponentsWithStats(src_bin)
dst = cv2.cvtColor(src, cv2.COLOR_GRAY2BGR)

print(cnt)
print(labels)
print(labels)
print(stats)
print(centroids)

for i in range(1,cnt):
    (x, y, w, h, area) = stats[i]
    # 픽셀 10개 미만이면 bounding box 예외처리
    if area < 10:
        continue
    cv2.rectangle(dst, (x, y, w, h), (0, 0, 255), 2)


cv2.imshow('src',src)
cv2.imshow('src_bin',src_bin)
cv2.imshow('dst',dst)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#ㅎ ㅏ...안찾아진다 공
import cv2
import numpy as np
import math

def setLabel(img, pts, label):
    (x, y, w, h) = cv2.boundingRect(pts)
    pt1 = (x,y)
    pt2 = (x + w, y + h)
    cv2.rectangle(img, pt1, pt2, (0, 255, 0), 1)
    cv2.putText(img, label, pt1, cv2.FONT_HERSHEY_PLAIN, 1, (0, 0, 255))
    
img = cv2.imread('basketball.jpg', cv2.IMREAD_COLOR)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_thr = cv2.adaptiveThreshold(img_gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,15,1)
cv2.imshow('thr', img_thr)

contours, _ = cv2.findContours(img_thr, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

for pts in contours:
    #if cv2.contourArea(pts) < 10:  #  너무 작으면 무시
     #   continue
    
    approx = cv2.approxPolyDP(pts, cv2.arcLength(pts, True)*0.05, True)
    vtc = len(approx)
    
    if cv2.isContourConvex(pts) and vtc != 3 and vtc != 4:
        length = cv2.arcLength(pts, True)
        area = cv2.contourArea(pts)
        ratio = 4 * math.pi * area / (length * length)
        
        if ratio > 0.8:
            setLabel(img, pts, 'ball')
            
cv2.imshow('img', img)
cv2.waitKey()
cv2.destroyAllWindows()
```

## 이미지변환

affine transform

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('banana.jpg')
rows, cols = img.shape[:2]

#변환 전후 각 좌표 3개 생성
pts1 = np.float32([[100,50],[200,100],[100,200]])
pts2 = np.float32([[80,70],[210,60],[250,120]])

#변환 전 좌표 이미지에 표시
cv2.circle(img, (100,50), 5, (255,0,0), -1)
cv2.circle(img, (200,100), 5, (0,255,0), -1)
cv2.circle(img, (100,200), 5, (255,0), -1)

#변환행렬계산
mtrx = cv2.getAffineTransform(pts1,pts2)

#어떤 변환 적용
dst = cv2.warpAffine(img, mtrx, (int(cols*1.5), rows))

cv2.imshow('img',img)
cv2.imshow('dst', dst)
cv2.waitKey()
cv2.destroyAllWindows()
```

perspective transform

```python
import cv2

img = cv2.imread('banana.jpg')
rows, cols = img.shape[:2]

#원본4개의점, 결과4개의점
pts1 = np.float32([[0,0],[0,rows],[cols,0],[cols,rows]])
pts2 = np.float32([[100,50],[10,rows-50],[cols-100,50],[cols-10,rows-50]])

#원본영상에 4개의 점 표시
cv2.circle(img, (0,0), 10, (255,0,0), -1)
cv2.circle(img, (0,rows), 10, (0,255,0),-1)
cv2.circle(img, (cols,0), 10, (0, 0, 255),  -1)
cv2.circle(img, (cols,rows), 10, (0, 255, 255), -1)

#원근 변환 행렬 계산
mtrx = cv2.getPerspectiveTransform(pts1, pts2)

#원근변환적용
dst = cv2.warpPerspective(img, mtrx, (cols, rows))

cv2.imshow('img', img)
cv2.imshow('perspective', dst)

cv2.waitKey()
cv2.destroyAllWindows()
```

실시간 영상에 엣지 표시하기

```python
import cv2
import numpy as np

cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH,320)
cap.set(cv2.CAP_PROP_FRAME_WIDTH,240)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
        
    frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    edge = cv2.Canny(frame_gray, 80, 127)
    
    edge1 = cv2.bitwise_not(edge)
    edge1 = cv2.cvtColor(edge1,cv2.COLOR_GRAY2BGR)
    edge = cv2.cvtColor(edge, cv2.COLOR_GRAY2BGR)
    
    img1 = cv2.bitwise_and(frame, edge1)
    img2 = cv2.bitwise_and(frame, edge)
    
    merged = np.hstack((frame, img2, img1))
    
    cv2.imshow('edge Camera', merged)
        
    if cv2.waitKey(1)==27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

def oil_painting_effect(src, radius, levels):
    dst = np.zeros(src.shape, src.dtype)
    gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
    
    for i in range(src.shape[0]):
        print(i)
        for j in range(src.shape[1]):
            hist = np.zeros(levels)# 이웃픽셀들 히스토그램 저장 변수
            bgr_sum = np.zeros((levels,3))# 이웃 픽셀들 컬러값 합 저장
            
            # 이웃 픽셀 모두 방문하며 히스토그램과 컬러합계산
            for u in range(max(i-radius,0), min((i+radius+1), src.shape[0])):
                for v in range(max(j-radius,0), min(j+radius+1, src.shape[1])):
                    #해당 히스토그램상자 인덱스 계산
                    level = gray[u,v] * levels//256
                    hist[level] += 1# 히스토그램 누적
                    bgr_sum[level] += src[u,v]
                    
            max_idx = np.argmax(hist) # 히스토그램 값 가장 큰 인덱스
            # 가장 많은 픽셀 속한 히스토그램 구간에 있는 픽셀 평균 컬러 값
            dst[i,j] = bgr_sum[max_idx]//hist[max_idx]
    
    return dst    
            
img = cv2.imread('red-and-green-leavs-of-autumn-jpg.jpg')

radius, levels = 4, 8
effect = oil_painting_effect(img, radius, levels)
cv2.imshow('img', img)
cv2.imshow('effect',effect)
cv2.waitKey()
cv2.destroyAllWindows()
```
