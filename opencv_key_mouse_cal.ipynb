영상을 grayscale로 변환 후 i를 누르면 영상의 밝기가 반대로 되는 프로그램을 개발하시오

```python
img=cv2.imread('banana.jpg', cv2.IMREAD_GRAYSCALE)
cv2.imshow('image',img)

while True:
    keycode = cv2.waitKey()
    if keycode == ord('i'):
        img = -img
        cv2.imshow('image',img)
    elif keycode == 27:
        break
            
cv2.destroyAllWindows()
```

컬러영상을 입력받은 후 r을 누르면 BRG영상이 RGB로 바뀌고 다시 r을 누르면 BRG 영상으로 바뀌는 코드를 작성하시오

```python
import cv2
img = cv2.imread('banana.jpg', cv2.IMREAD_COLOR)
cv2.imshow('Image', img)

while True:
    keycode=cv2.waitKey()
    if keycode == ord('r'):
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        cv2.imshow('Image', img)
    elif keycode == ord('i'):
        img = -img
        cv2.imshow('Image', img)
    elif keycode == 27:
        break

cv2.destroyAllWindows()
```

```python
#내가한거
import cv2
import numpy as np

def B_level_change(pos):
    value = pos
    if value>=255:
        value = 255
    img[:,:,0] = value
    
    cv2.imshow('image', img)
def G_level_change(pos):
    value = pos
    if value>=255:
        value = 255
    img[:,:,1] = value
    
    cv2.imshow('image', img)

def R_level_change(pos):
    value = pos
    if value>=255:
        value = 255
    img[:,:,2] = value
    
    cv2.imshow('image', img)

img = np.zeros((380, 640), np.uint8)
img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)

cv2.namedWindow('image')
cv2.createTrackbar('R level', 'image', 0, 255, R_level_change)
cv2.createTrackbar('G level', 'image', 0, 255, G_level_change)
cv2.createTrackbar('B level', 'image', 0, 255, B_level_change)

cv2.imshow('image', img)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import sys
import cv2
import numpy as np

star = cv2.imread('star.jpg', cv2.IMREAD_GRAYSCALE)

h, w = star.shape

src2 = np.zeros((h, w), dtype = np.uint8)

for x in range(int(w/2)):
    for y in range(h):
        src2[y, x] = 255
        
cv2.imshow('src2', src2)

bit_and = cv2.bitwise_and(star, src2)
bit_or = cv2.bitwise_or(star, src2)
bit_xor = cv2.bitwise_xor(star, src2)
bit_not = cv2.bitwise_not(star)

cv2.imshow('star', star)
cv2.imshow('src2_after_bit', src2)
cv2.imshow('bit_and', bit_and)
cv2.imshow('bit_or', bit_or)
cv2.imshow('bit_xor', bit_xor)
cv2.imshow('bit_not', bit_not)

cv2.waitKey()

cv2.destroyAllWindows()
```

### Filtering

numpy이용해 3x3 median filter 이용해 grayscale영상을 blurring 해보세요

```python
import cv2
import numpy as np

src = cv2.imread('banana.jpg', cv2.IMREAD_GRAYSCALE)
kernel = np.full((3, 3), 1/9)

filtered_banana = cv2.filter2D(src, -1, kernel, anchor=(-1, -1))

cv2.imshow('banana', src)
cv2.imshow('filtered', filtered_banana)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#kernel 크기 조정으로 blur정도 조절하기
import numpy as np
import cv2

def nothing(x):
    pass

src = cv2.imread('fruit.jpg', cv2.IMREAD_GRAYSCALE)
cv2.namedWindow('frame')#트랙바 생성할 윈도우 생성
cv2.imshow('frame', src)
cv2.createTrackbar('Blur', 'frame', 0, 20, nothing) #트랙바값변경시마다호출할콜백함수 nothing

while True:
    
    key = cv2.waitKey(1)
    if key == 27:
        break
    
    b = cv2.getTrackbarPos('Blur', 'frame') #트랙바값가져오고
    k = b*2+1 #홀수로 지정->kernel크기로 설정
    
    dst1 = cv2.blur(src, (k,k)) #blur함수
    cv2.imshow('frame', dst1)
    

cv2.destroyAllWindows()
```

```python
#mode 1: line, 2:rectangle, 3:circle, 4:free

import cv2
import numpy as np

x0, y0 = 0, 0
draw = False
mode = 1
img = np.zeros((480, 640, 3), np.uint8)
img0 = np.copy(img)

def on_mouse(event, x, y, flags, param):
    global x0, y0, draw, mode, img, img0
    
    if event == cv2.EVENT_LBUTTONDOWN:
        draw = True
        x0, y0 = x, y
    elif draw and event == cv2.EVENT_MOUSEMOVE:
        if mode < 4:np.copyto(img, img0)
        if mode == 1:
            cv2.line(img, (x0, y0), (x, y), (0, 0, 255), 2)
        elif mode == 2:
            cv2.rectangle(img, (x0, y0), (x, y), (0, 255, 0), 2)
        elif mode == 3:
            r = np.sqrt((x-x0)*(x-x0)+(y-y0)*(y-y0))
            cv2.circle(img, (x0, y0), int(r), (0, 255, 255), 2)
        elif mode == 4:
            cv2.line(img, (x0, y0), (x,y), (255, 0, 255), 2)
    elif event == cv2.EVENT_LBUTTONUP:
        draw = False
        np.copyto(img0, img)
        
cv2.namedWindow('img')
cv2.setMouseCallback('img', on_mouse)

while True:
    cv2.imshow('img', img)
    key = cv2.waitKey(30)
    if key > 48 and key < 53: mode = key - 48
    elif key == 27: break

cv2.destroyAllWindows()
```

```python
#multi trackbars
import cv2 as cv
import numpy as np

def onChangeB(val):
    print('B:', val)
    img[:, :, 0] = val
    cv.imshow('img', img)
def onChangeG(val):
    print('G:', val)
    img[:, :, 1] = val
    cv.imshow('img', img)
def onChangeR(val):
    print('R:', val)
    img[:, :, 2] = val
    cv.imshow('img', img)

init_val = 128
img = np.full((480, 640, 3), init_val, np.uint8)
cv.namedWindow('img')
cv.createTrackbar('R', 'img', init_val, 255, onChangeR)
cv.createTrackbar('G', 'img', init_val, 255, onChangeG)
cv.createTrackbar('B', 'img', init_val, 255, onChangeB)

cv.imshow('img',img)
cv.waitKey()
cv.destroyAllWindows()
```
